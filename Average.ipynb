```
def calculate_average():
    """Calculates the average of four grades and checks for approval."""
    # Get four grades from the user
    grade1 = int(input("Enter the first grade: "))
    grade2 = int(input("Enter the second grade: "))
    grade3 = int(input("Enter the third grade: "))
    grade4 = int(input("Enter the fourth grade: "))

    # Calculate the average
    average = (grade1 + grade2 + grade3 + grade4) / 4

    # Check for approval based on the average
    check_approval(average)

def check_approval(average):
    """Checks if a student is approved or reproved based on their average grade.

    Args:
        average (float): The calculated average grade.
    """
    if average >= 8:
        print(f"Approved with average: {average:.2f}\n")
    else:
        print(f"Reproved with average: {average:.2f}\n")

# Main program execution
def main():
    """Main function to get the number of students and process their grades."""
    num_students = int(input("Enter the number of students to be evaluated: "))

    # Loop through each student
    for student_number in range(1, num_students + 1):
        print(f"Student {student_number}:\n")
        calculate_average()

# Call the main function to start the program
if __name__ == "__main__":
    main()
```

### Explicação do Código Python:

*   **`def calculate_average():`**: Esta função solicita quatro notas ao usuário, calcula a média e chama a função `check_approval` para determinar o status do aluno.
*   **`def check_approval(average):`**: Esta função recebe a média como argumento e imprime se o aluno foi 'Approved' (aprovado) ou 'Reproved' (reprovado), junto com a média.
*   **`def main():`**: Esta é a função principal que solicita a quantidade de alunos a serem avaliados e, em seguida, itera chamando `calculate_average` para cada aluno.
*   **`if __name__ == "__main__":`**: Esta é uma construção padrão em Python que garante que a função `main()` seja chamada apenas quando o script é executado diretamente (e não quando é importado como um módulo em outro script).
